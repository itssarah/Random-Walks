## Simulating 2D Random Walk

#### Importing important libraries


```python
import numpy as np
np.random.seed(1234)
import matplotlib.pyplot as plt
import matplotlib.animation as animation
```

#### Defining a 2D Random Walk

```python
# Define parameters for the walk
dims = 2
step_n = 10000
step_set = [-1, 0, 1]
origin = np.zeros((1,dims))
# Simulate steps in 2D
step_shape = (step_n,dims)
steps = np.random.choice(a=step_set, size=step_shape)
path = np.concatenate([origin, steps]).cumsum(0)
start = path[:1]
stop = path[-1:]
```

#### Plotting

```python
# Plot the path
fig = plt.figure(figsize=(10,10),dpi=200)
ax = fig.add_subplot(111)
plt.gca().set_aspect('equal', adjustable='box')
ax.scatter(path[:,0], path[:,1],c="blue",alpha=0.25,s=0.05);
ax.plot(path[:,0], path[:,1],c="blue",alpha=0.5,lw=0.25,ls='-');
ax.plot(start[:,0], start[:,1],c="red", marker='+')
ax.plot(stop[:,0], stop[:,1],c="black", marker='o')
plt.title('2D Random Walk')
plt.tight_layout(pad=0)
```

#### Animated Plot

```python
# Define parameters for the walk

dims = 2
step_n = 50
step_set = [-1, 0, 1]
origin = np.zeros((1,dims))
# Simulate steps in 2D
step_shape = (step_n,dims)
steps = np.random.choice(a=step_set, size=step_shape)
path = np.concatenate([origin, steps]).cumsum(0)
start = path[:1]
stop = path[-1:]


# Plot the path
fig = plt.figure(figsize=(8,8),dpi=200)
ax = plt.axes(xlim=(np.min(path[:,0]) - 0.5, np.max(path[:,0]) + 0.5), ylim=(np.min(path[:,1]) - 0.5, np.max(path[:,1]) + 0.5)) 
ax.plot(start[:,0], start[:,1],c="blue", marker='+')

line, = ax.plot([], [], c="blue",alpha=0.5,lw=0.25,ls='-')
scat = ax.scatter(x=[path[0,0]], y=[path[0,1]], c="blue",alpha=0.25,s=0.7 )


plt.title('2D Random Walk')
plt.tight_layout(pad=0)

def init(): 
    # creating an empty plot/frame 
    line.set_data([], []) 
    return line, 

# lists to store x and y axis points 
xdata, ydata = [], []

def animate(i):
    
    # appending new points to x, y axes points list 
    xdata.append(path[i,0]) 
    ydata.append(path[i,1]) 
    x = path[:i,0]
    y = path[:i,1]
    scat.set_offsets(np.c_[x,y])
    line.set_data(xdata, ydata)
    return line,  scat,




# call the animator	 
anim = animation.FuncAnimation(fig, animate, init_func=init, frames=step_n, interval=20, blit=True)
f = r"./rd_2d.gif" 
writergif = animation.PillowWriter(fps=50) 
anim.save(f, writer=writergif)
plt.savefig('./random_walk_2d.png',dpi=250);
```

#### Multiple 2d Walks

```python
# Define parameters for the walk
dims = 2
step_n = 50
step_set = [-1, 0, 1]
origin = np.zeros((1,dims))
# Simulate steps in 2D
step_shape = (step_n,dims)
steps = np.random.choice(a=step_set, size=step_shape)
path = np.concatenate([ steps]).cumsum(0)
start = path[:1]
stop = path[-1:]

step_shape = (step_n,dims)
steps1 = np.random.choice(a=step_set, size=step_shape)
path1 = np.concatenate([steps1]).cumsum(0)
start1 = path1[:1]
stop1 = path1[-1:]

step_shape = (step_n,dims)
steps2 = np.random.choice(a=step_set, size=step_shape)
path2 = np.concatenate([ steps2]).cumsum(0)
start2 = path2[:1]
stop2 = path2[-1:]

# Plot the path
fig = plt.figure(figsize=(8,8),dpi=200)
ax = plt.axes(xlim=(min(np.min(path[:,0]),np.min(path1[:,0]),np.min(path2[:,0])) - 0.5,  
                    max(np.max(path[:,0]),np.max(path1[:,0]),np.max(path2[:,0])) + 0.5), 
              ylim=(min(np.min(path[:,1]),np.min(path1[:,1]),np.min(path2[:,1])) - 0.5, 
                    max(np.max(path[:,1]),np.max(path1[:,1]),np.max(path2[:,1])) + 0.5)) 

ax.plot(start[:,0], start[:,1],c="#00bfbf", marker='+')
ax.plot(start1[:,0], start1[:,1],c="#bf00bf", marker='+')
ax.plot(start2[:,0], start2[:,1],c="#0000ff", marker='+')



line, = ax.plot([], [], c="#00bfbf",alpha=0.5,lw=0.25,ls='-')
scat = ax.scatter(x=[path[0,0]], y=[path[0,1]], c="#00bfbf",alpha=0.25,s=0.7 )

line1, = ax.plot([], [], c="#bf00bf",alpha=0.5,lw=0.25,ls='-')
scat1 = ax.scatter(x=[path1[0,0]], y=[path1[0,1]], c="#bf00bf",alpha=0.25,s=0.7 )

line2, = ax.plot([], [], c="#0000ff",alpha=0.5,lw=0.25,ls='-')
scat2 = ax.scatter(x=[path2[0,0]], y=[path2[0,1]], c="#0000ff",alpha=0.25,s=0.7 )


plt.title('2D Random Walk')
plt.tight_layout(pad=0)

def init(): 
    # creating an empty plot/frame 
    line.set_data([], []) 
    return line, 

# lists to store x and y axis points 
xdata, ydata = [], []
xdata1, ydata1 = [], []
xdata2, ydata2 = [], []

def animate(i):
    
    # appending new points to x, y axes points list 
    xdata.append(path[i,0]) 
    ydata.append(path[i,1]) 
    x = path[:i,0]
    y = path[:i,1]
    scat.set_offsets(np.c_[x,y])
    line.set_data(xdata, ydata)
    
    xdata1.append(path1[i,0]) 
    ydata1.append(path1[i,1]) 
    x1 = path1[:i,0]
    y1 = path1[:i,1]
    scat1.set_offsets(np.c_[x1,y1])
    line1.set_data(xdata1, ydata1)
    
    xdata2.append(path2[i,0]) 
    ydata2.append(path2[i,1]) 
    x2 = path2[:i,0]
    y2 = path2[:i,1]
    scat2.set_offsets(np.c_[x2,y2])
    line2.set_data(xdata2, ydata2)
    
    return line,  scat, line1,  scat1, line2,  scat2,




# call the animator	 
anim = animation.FuncAnimation(fig, animate, init_func=init, frames=step_n , interval=20, blit=True)
f = r"./rd_2d_multiple.gif" 
writergif = animation.PillowWriter(fps=50) 
anim.save(f, writer=writergif)


plt.savefig('./random_walk_2d.png',dpi=250);
```
